```python
year = 1980
county_data = []
while year <= 2018:
    file = pd.read_csv(f"daily_aqi_by_county_{year}.csv")
    filedf = pd.DataFrame(file)    
    datedf = filedf["Date"].str.split("-",expand = True)
    datedf.columns = ["Year","Month","Day"]
    filedf["Year"] = datedf["Year"]
    filedf["Month"] = datedf["Month"]
    filedf["Day"] = datedf["Day"]
    file2df = filedf.groupby(["county Name", "Year"])
    countydatedf = file2df.mean().unstack(level = 1)
    countydf = countydatedf["AQI"]
    county_data.append(countydf)
    year += 1
```

```python
statedf = pd.DataFrame(county_data)
```
