```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
file2 = pd.read_csv(f"../almostfinishedaqidata_smile_/aqi_monthly_county/2015countyaqi.csv")
file2df = pd.DataFrame(file2)
file = pd.read_csv("2005_final_df.csv")
df = pd.DataFrame(file)
x = df["Longitude"]
y = df["Latitude"]
z = df["01"]
cmap = sns.cubehelix_palette(as_cmap=True)

f, ax = plt.subplots()
points = ax.scatter(x, y, c=z, s=50, cmap=cmap, alpha = .5)
plt.title("AQI by county January, 2005")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
f.colorbar(points)
plt.show()
```
