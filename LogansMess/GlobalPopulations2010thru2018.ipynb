```python
import json
import csv
import pandas as pd
from pprint import pprint as pprint
import requests
from config import censusapi, gmapsapi
import gmaps
import citipy
gmaps.configure(api_key = gmapsapi)
```

```python
data = pd.read_csv("2010globalpopulation.txt")
df = pd.DataFrame(data)
uniquecountries = df["NAME"].unique()
totalpopbycountrydf = pd.DataFrame(uniquecountries, columns = {"Country Name"})
```

```python
population = 0
negater = 0

years = []
for x in range(9):
    year = 2010 + x
    years.append(year)
```

```python
for year in years:
    dataloop = pd.read_csv(f"{year}globalpopulation.txt")
    dfloop = pd.DataFrame(dataloop)
    totalpopulation = []
    for i in range(len(df)):
        population += dfloop["POP"][i]
        if df["AGE"][i] == 100:
            countrypop = population - negater
            negater = population
            totalpopulation.append(countrypop)

    totalpopbycountrydf[f"{year} Total Population"] = totalpopulation
```

```python
countriesdata = pd.read_csv("countrieslatlon.txt")
latitude = []
longitude = []
countrycode = []
countryname = []
for i in range(len(countriesdata)):
    if countriesdata["Country"][i] in uniquecountries:
        countrycode.append(countriesdata["ISO 3166 Country Code"][i])
        countryname.append(countriesdata["Country"][i])
        latitude.append(countriesdata["Latitude"][i])
        longitude.append(countriesdata["Longitude"][i])
```

```python
countrydf = pd.DataFrame(countryname,columns = {"Country Name"},index = countrycode)
countrydf["Latitude"] = latitude
countrydf["Longitude"] = longitude
```

```python
countrydf.merge(totalpopbycountrydf)
```
