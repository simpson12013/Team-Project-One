```python
state_solar = requests.get(f'http://api.eia.gov/category/?api_key={api_key}&category_id=40382').json()
series_id = []
state_solar_info = []
df_list = []
childseries = state_solar['category']['childseries']
for state in range(len(childseries)):
    series = childseries[state]["series_id"]
    series_id.append(series)
for series in series_id:
    state_info = requests.get(f'http://api.eia.gov/series/?api_key={api_key}&series_id={series}').json()
    state_solar_info.append(state_info)
for i in range(len(state_solar_info)):
    name = state_solar_info[i]['series'][0]['geography']
    name = pd.DataFrame(state_solar_info[i]['series'][0]['data'][0])
    df_list.append(name)
    
pprint (state_solar_info)
    
```
